1.Email Validator

```python
email = input()

valid = False
if "@" in email:
    if email.endswith(".com") or email.endswith(".org"):
        valid = True

print("Valid Email" if valid else "Invalid Email")
```

2️⃣ Word Frequency Counter

```python
text = input().lower()
words = text.split()

freq = {}
for i in words:
    freq[i] = freq.get(i, 0) + 1

print(freq)
```

3️⃣ Password Strength Checker

```python
pwd = input()

has_upper = any(ch.isupper() for ch in pwd)
has_lower = any(ch.islower() for ch in pwd)
has_digit = any(ch.isdigit() for ch in pwd)
has_special = any(not ch.isalnum() for ch in pwd)

if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

4️⃣ Username Generator

```python
import random

full = input().strip().lower().split()
first, last = full[0], full[-1]

uname = first[:3] + last[-3:] 

# random 2-digit number using loop only
num = ""
for _ in range(2):
    num += str(random.randint(0, 9))

print(uname + num)
```

5️⃣ SMS Word Counter

```python
msg = input()

words = msg.split()
chars = len(msg)

status = "Within Limit" if chars <= 160 else "Exceeded Limit"

print(f"Words = {len(words)}, Characters = {chars}, Status = {status}")
```

6️⃣ Remove Duplicates

```python
names = input().split(",")
clean = []

for n in names:
    n = n.strip()
    if n not in clean:
        clean.append(n)

print(", ".join(clean))
```

7️⃣ Suspicious Word Detector

```python
text = input().lower()
banned = ['spam', 'scam', 'fake']

flag = False
for b in banned:
    if b in text:
        flag = True
        break

if flag:
    print("Warning: Suspicious content detected")
else:
    print("Clean")
#
```

8️⃣ Invoice Code Formatter

```python
n = int(input())

# Pad manually using loops
num_str = str(n)
while len(num_str) < 5:
    num_str = "0" + num_str

print("INV" + num_str)
```
